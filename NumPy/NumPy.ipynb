```python
#calcular a norma entre as duas funções
norma = np.sqrt(np.sum(np.power(Moscow-y,2)))
norma
```

```python
#calculo da norma por uma função especifica do np
norma2 = np.linalg.norm(Moscow-y)
norma2
```

para fazer a atribuição de uma array em outra e depois poder modifica-la sem o risco de alterar as duas é preciso usar a função np.copy()

```python
# calculo dos coeficientes a e b
X = datas
Y = Moscow
n = np.size(Moscow)

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - (np.sum(X))**2)
b = np.mean(Y) - a*np.mean(X)

y = a*x + b
```

```python
np.linalg.norm(Moscow-y)
```

```python
plt.plot(datas,Moscow,'black')
plt.plot(x,y)
plt.show()
```

# Aula 4 (números aleatórios)

```python
#gerar números aleatórios inteiros
np.random.randint(low=40,high=100,size=100)
```

```python
#gerar números aleatórios do tipo float
np.random.uniform(low=0,high=2,size=100)
```

```python
'''
Os números gerados atravez da função random não são completamente aleatórios.
Eles são pseudo-aleatórios e são gerados a partir de uma semente que pode ser definida com a função .seed(), desse modo é possível
conseguir sempre os mesmos números.
A semente deve ficar na mesma celula do gerador de números aleatórios
'''
np.random.seed(16)
np.random.uniform(low=0.1,high=0.9,size=100)
```

```python
np.random.seed(84)
coef_angular = np.random.uniform(low=0.1,high=0.9,size=100)
norma2 = np.array([])
for i in range(100):
  norma2 = np.append(norma2,np.linalg.norm(Moscow-(coef_angular[i]*x+b)))
```

```python
norma2
```

```python
dados = np.column_stack([norma2,coef_angular])
```
