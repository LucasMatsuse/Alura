```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
dados_hospedagem = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_hospedagem.json')
dados_hospedagem
```

```python
dados_hospedagem = pd.json_normalize(dados_hospedagem['info_moveis'])
```

```python
dados_hospedagem
```

```python
dt_datas = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/moveis_disponiveis.json')
dt_datas
```

```python
colunas = list(dados_hospedagem.columns)
```

```python
dados_hospedagem = dados_hospedagem.explode(colunas[3:])
```

```python
dados_hospedagem.reset_index(inplace=True,drop=True)
dados_hospedagem.head()
```

```python
dados_hospedagem.info()
```

```python
col_num = ['max_hospedes','quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados_hospedagem[col_num] = dados_hospedagem[col_num].astype(np.int64)
dados_hospedagem.info()
```

```python
dados_hospedagem['avaliacao_geral'] = dados_hospedagem['avaliacao_geral'].astype(np.float64)
dados_hospedagem.info()
```

```python
col_cifrao = ['taxa_deposito','taxa_limpeza','preco']
```

```python
for i in col_cifrao:
  dados_hospedagem[i] = dados_hospedagem[i].apply(lambda x: x.replace('$','').replace(',','').strip())
  dados_hospedagem[i] = dados_hospedagem[i].astype(np.float64)
```

```python
'''
A alternativa a coluna acime é:
dados_hospedagem[['taxa_deposito','taxa_limpeza','preco']] = dados_hospedagem[['taxa_deposito','taxa_limpeza','preco']].applymap(lambda x: x.replace('$','').replace(',','').strip())
dados_hospedagem[['taxa_deposito','taxa_limpeza','preco']] = dados_hospedagem[['taxa_deposito','taxa_limpeza','preco']].astype(np.float64)
'''
```

```python
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ',regex=True)
```
