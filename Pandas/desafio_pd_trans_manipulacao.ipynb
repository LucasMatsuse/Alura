```python
import pandas as pd
import numpy as np
```

#Projeto 1: Vendas online

```python
dados_clientes = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json')
```

```python
dados_clientes = pd.json_normalize(dados_clientes['dados_vendas'])
dados_clientes.head()
```

```python
dados_clientes = dados_clientes.explode(['Cliente','Valor da compra'])
dados_clientes = dados_clientes.reset_index(drop=True)
```

```python
dados_clientes['Valor da compra'] = dados_clientes['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','.').strip())
```

```python
dados_clientes['Valor da compra'] = dados_clientes['Valor da compra'].astype(np.float64)
```

```python
dados_clientes.head()
```

```python
dados_clientes['Cliente'] = dados_clientes['Cliente'].str.replace('[^a-zA-Z\ ]','',regex=True)
dados_clientes['Cliente'] = dados_clientes['Cliente'].apply(lambda x: x.lower())
dados_clientes.info()
```

```python
dados_clientes
```

```python
dados_clientes['Data de venda'] = pd.to_datetime(dados_clientes['Data de venda'],format='%d/%m/%Y')
```

```python
compra_total_clinete = dados_clientes.groupby('Cliente')['Valor da compra'].sum()
```

```python
compra_total_clinete
```

# Projeto 2 : administração de condomínios

```python
dados_locacao = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json')
```

```python
dados_locacao = pd.json_normalize(dados_locacao['dados_locacao'])
```

```python
dados_locacao.head()
```

```python
dados_locacao = dados_locacao.explode(['datas_combinadas_pagamento','datas_de_pagamento','valor_aluguel'])
dados_locacao.reset_index(inplace=True,drop=True)
```

```python
dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].apply(lambda x: x.replace('$','').replace(',','.').replace(' reais','').strip())
dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].astype(np.float64)
```

```python
dados_locacao['apartamento'] = dados_locacao['apartamento'].str.replace('\(blocoAP\)','',regex=True)
```

```python
dados_locacao['datas_combinadas_pagamento'] = pd.to_datetime(dados_locacao['datas_combinadas_pagamento'],format='%d/%m/%Y')
```

```python
dados_locacao['datas_de_pagamento'] = pd.to_datetime(dados_locacao['datas_de_pagamento'],format='%d/%m/%Y')
```

```python
dias_de_atraso = dados_locacao['datas_de_pagamento'] - dados_locacao['datas_combinadas_pagamento']
```

```python
dados_locacao['dias_de_atraso'] = dias_de_atraso.dt.days
```

```python
media_atraso = dados_locacao.groupby('apartamento')['dias_de_atraso'].mean()
```
