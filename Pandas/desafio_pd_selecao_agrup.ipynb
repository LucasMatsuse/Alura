```python
filtro4 = dados['Emissão / Remoção / Bunker'] == 'Emissão'
dados_d2 = dados[filtro4]
dados_d2.head()
```

```python
dados_d2.drop(columns='Emissão / Remoção / Bunker',inplace=True)
```

```python
dados_d2.groupby('Nível 1 - Setor').groups
```

```python
agro = dados_d2.groupby('Nível 1 - Setor').get_group('Agropecuária')
agro
```

```python
dados_d2.groupby('Nível 1 - Setor').mean()[2021]
```

```python
'''
Mudar o formato do DataFrame para passar a informação dos anos para linhas
1-selecionar as colunas que se manterão no DT
2-selecionar as colunas que irão virar linhas
3-transformar as duas informações em listas
4-usar o .melt()
'''
colunas = list(dados_d2.loc[:,'Nível 1 - Setor':'Produto'].columns)
colunas
```

```python
anos = list(dados_d2.loc[:,1970:2021].columns)
anos
```

```python
dados_d2_ano = dados_d2.melt(id_vars=colunas,value_vars=anos,var_name='Ano',value_name='Emissão')
dados_d2_ano
```

```python
emissao_total_setor = dados_d2_ano.groupby('Nível 1 - Setor')[['Emissão']].sum().sort_values('Emissão',ascending=False)
```

```python
emissao_total_setor.plot(kind='barh',figsize=(14,10),ylabel='Setor',xlabel='Emissão(x 10¹¹)')
```
